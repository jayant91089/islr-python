```python
yoe = np.random.uniform(low=10,high=22, size=30)
```

```python
from scipy.special import expit
from scipy.optimize import curve_fit
from functools import partial
```

```python
income = 80*expit(yoe-16)+18+np.random.normal(size=yoe.shape[0], scale = 5)
```

```python
def func(x,a,b,c):
    return a*expit(x-b)+c
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(14,6))
axes = axes.flatten()
sb.scatterplot(x=yoe,y=income,ax=axes[0],facecolors="r")
popt,pcov = curve_fit(func,yoe,income,p0=[50,10,12])
func_opt=partial(func,a=popt[0],b=popt[1],c=popt[2])
sb.scatterplot(x=yoe,y=income,ax=axes[1],facecolors="r")
sb.lineplot(x=yoe,y=func_opt(yoe),ax=axes[1],color="b")
for i in range(len(yoe))
# sb.regplot(x=yoe,y=income,order=3,ax=axes[1],ci=None,scatter_kws={"facecolors":"r","ec":"r"},line_kws={"color":"b"})
```
